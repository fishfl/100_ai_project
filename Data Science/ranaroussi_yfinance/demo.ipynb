```python
import yfinance as yf

bili = yf.Ticker("BILI")

# get all stock info
bili.info
```

```python
# get historical market data
hist = bili.history(period="1mo")
hist
```

```python
# show news
# 显示新闻
bili.news
```

```python
# show meta information about the history (requires history() to be called first)
# 显示有关历史记录的元信息（需要首先调用history（））
bili.history_metadata
# show actions (dividends, splits, capital gains)
# 显示操作（股息，拆分，资本收益）
bili.actions
bili.dividends
bili.splits
bili.capital_gains  # only for mutual funds & etfs
# show share count
# 显示股票数量
bili.get_shares_full(start="2022-01-01", end=None)
# show financials:
# 显示财务状况：
bili.calendar
bili.sec_filings
# - income statement
# 收入表
bili.income_stmt
bili.quarterly_income_stmt
# - balance sheet
# 资产负债表
bili.balance_sheet
bili.quarterly_balance_sheet
# - cash flow statement
# 现金流量表
bili.cashflow
bili.quarterly_cashflow
# see `Ticker.get_income_stmt()` for more options

# show holders
# 显示持有人
bili.major_holders
bili.institutional_holders
bili.mutualfund_holders
bili.insider_transactions
bili.insider_purchases
bili.insider_roster_holders

bili.sustainability

# show recommendations
# 显示推荐
bili.recommendations
bili.recommendations_summary
bili.upgrades_downgrades
# show analysts data
# 显示分析师数据
bili.analyst_price_targets
bili.earnings_estimate
bili.revenue_estimate
bili.earnings_history
bili.eps_trend
bili.eps_revisions
bili.growth_estimates
# Show future and historic earnings dates, returns at most next 4 quarters and last 8 quarters by default.
# Note: If more are needed use msft.get_earnings_dates(limit=XX) with increased limit argument.
# 显示未来和历史收益日期，默认情况下最多返回下一个季度和最近的8个季度。
bili.earnings_dates
# show ISIN code - *experimental*
# ISIN = International Securities Identification Number
# 显示ISIN代码- *实验性*
bili.isin
# show options expirations
# 显示期权到期
bili.options
# get option chain for specific expiration
# 获取特定到期的期权链
opt = bili.option_chain('2024-11-29')
# data available via: opt.calls, opt.puts
```
